```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

sns.set_style('whitegrid')
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer_data = load_breast_cancer()
```

```python
cancer_data.keys()
```

```python
print(cancer_data['DESCR'])
```

```python
print(cancer_data['feature_names'])
```

```python
df = pd.DataFrame(cancer_data['data'], columns=cancer_data['feature_names'])
```

```python
df.head()
```

```python
cancer_data['target']
```

```python
cancer_data['target_names']
```

```python
# pca visualization
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(df)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
pca.fit(scaler.transform(df))
```

```python
x_pca = pca.transform(scaler.transform(df))
```

```python
scaler.transform(df).shape
```

```python
x_pca.shape
```

```python
plt.figure(figsize=(8,6))
# get all the rows from column 0 and plot them againt all the rows from column 1
plt.scatter(x_pca[:,0], x_pca[:,1], c=cancer_data['target'], cmap='plasma')
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
```

```python
pca.components_
```

```python
df_comp = pd.DataFrame(pca.components_, columns=cancer_data['feature_names'])
```

```python
df_comp
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(df_comp, cmap='plasma')
```
